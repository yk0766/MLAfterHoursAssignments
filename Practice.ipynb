### 1. Bisection Method

We consider linear equations of the form $f(x) = mx+ c$, where $m$ is the slope of the line and $c$ is the y-intercept.
 For convenience, we assume that the domain of the function (the valid values of $x$) is restricted to some range.
 The root of an equation is the value of x for which $f(x)= 0$

 For example, here is a visualization of the equation $y = 2x + 5$ over the range $[0, 4]$


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 4, 100)
y = 2*x + 5

plt.plot(x, y)
plt.title("Linear Function")
plt.grid()
plt.show()
```

As you can see, the $y$ values vary from $5$ to $13$. We will choose a domain ($x$ values) such that there is a point where $y = 0$, or $y$ crosses the $x$-axis

As an aside, you should familiarize yourself with the algebraic and geometric view of a function:
- The line is the set of all points that satisfy the equation.
- Any point that satisfies the equation is on this line.

Algebraically, finding the root of the equation is finding the value of $x$ such that $f(x) = 0$. Geometrically, it is determining when the line crosses the $x$-axis. 

We usually solve for $x$ ($x = \frac{-c}{m}$) symbolically. On a computer, we want to solve this numerically, but finding an approximation and refining this value. 

The bisection method is one of the simplest numerical algorithms to find the
root of an equation.

Hereâ€™s how we proceed. We first find two values of x, say
$x_1$ and $x_2$, such that $f(x_1) < 0$ and $f(x_2) > 0$. 

For example, consider the following graph of $y = -2x + 3$.

A possible value 
of $x_1$ is `2`. Or `1.615`. Or, in fact, any number greater than `1.5` (which we know
just by looking at the graph, but in general, we would like to automate this process).

```python
x = np.linspace(-3, 3, 100)
y = -2*x + 3
plt.plot(x, y)
plt.title("Linear Function")
plt.grid()
plt.axhline(0, color='black', linewidth=1.5)
plt.axvline(0, color='black', linewidth=1.5)
plt.xlim(-3, 3)
```

The approach to determining a suitable $x_1$ and $x_2$ is to use `np.random.rand()`, which will
generate a random number between `0` and `1`. You will have to map it to the correct range, which 
needs a bit of experimentation. Try it out!

If you generate a single random number, you may or may not get a suitable value. If not, you will
have to keep generating random numbers, till your condition is satisfied.


The intuition behind the bisection method is pretty straightforward: if we have
two values of x for which the function is less than and greater than zero, then
there must be some value of x between these two values for which the function is
equal to zero. So we simply calculate $x = \frac{x_1+x_2}{2}$ and set either
$x_1$ or $x_2$ to this new value, after evaluating the function to check if it
is negative or positive for this value. We iterate till  $x_1$ equals $x_2$,
which is the root we seek.

### 2. Collatz Conjecture

For any integer $n$, we can perform the following operations repeatedly:
- if n is even, set $n = n // 2$.
- if n is odd,  set $n = 3n + 1$.

The conjecture states that n is eventually 1

There is no proof for this conjecture (hence, it's a conjecture, not a theorem). However, it has been tested for values of $n$ to many billions.
