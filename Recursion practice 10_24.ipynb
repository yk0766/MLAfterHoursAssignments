### 1. Find the sum of a list of integers

```python
lst = [2, 5, 76 ,34, 5]
```

```python
def getsum(lst):
    if len(lst)<1:
        return 0 
    else:
        res = lst[0] + getsum(lst[1:])
        return res
```

```python
print("The sum is",getsum(lst))
```

### 2. Find the length of a string

```python
def get_len(name):
    if len(name) < 1:
        return 0 
    else:
        lens = 1 + get_len(name[1:])
        return lens
```

### 3. â Find the max integer in a list of integers

```python
lst = [111, 5, 100 ,34, 99, 200]
```

```python
def getmax(lst):
    if len(lst)<2:
        return lst[0]
    else:
        if lst[0]>getmax(lst[1:]):
            maxn = lst[0]
            return maxn
        else:
            maxn = getmax(lst[1:])
            return maxn
```

### 4. Factorial n

```python
def fact(n):
    if n == 0:
        return 1
    else:
        res = n * fact(n-1)
    return res
```

```python
res = fact(5)
```

```python
print(res)
```

Find x^n

```python
def pow(x,n):
    if n == 0:
        return 1
    else:
        res = x * pow(x,n-1)
        return res
    
```

```python
pow(2,3)
```

### Find the nth Fibonacci number

```python
def fib(n):
    if n <= 1:
        return 0
    elif n== 2:
        return 1
    return (fib(n-2) + fib(n-1))
    
```

```python
fib(7)
```

### Determine if a string is a palindrome

```python
def check_pal(s):
    if len(s) <= 1:
        return True
    return (s[0] == s[-1] and check_pal(s[1:-1]))
    
```

```python
check_pal("civic")
```

### Count the number of times a character occurs in a string

```python
def count_char(ch,s):
    if len(s) < 1:
        return 0
    elif s[0] == ch:
        return (1 + count_char(ch,s[1:]))
    else:
        return (0 + count_char(ch,s[1:]))
```

```python
count_char('p','apparatus')
```

### Binary search

```python
def bin_search(n,lst):
    lst.sort()
    if len(lst) == 0:
        return False

    mid = ((len(lst)+1)//2)-1
    if lst[mid] > n:
        return bin_search(n,lst[:mid])
    elif lst[mid] < n:
        return bin_search(n,lst[mid+1:])
    else:
        return True
```

```python
num = 11
search = [1,3,5,7,9,11]
```

```python
bin_search(num, search)
```
