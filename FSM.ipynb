**FINITE STATE MACHINE**

**1. Rolling dice 6, 6 in row**

```python
def updatestate_6_6(currentstate, outcome):
    if((currentstate == "S") and (outcome == 6)): # "Start" state & Dice roll 6 -> "Intermediate"
        return "I"
    elif((currentstate == "I") and (outcome == 6)): # "Intermediate" state & Dice roll 6 -> "Finish"
        return "F"
    else:
        return "S" # "Other states & Dice roll = 1-5 -> "Start"
        
```

**2. Rolling dice 5, 6 in row**

```python
def updatestate_5_6(currentstate, outcome):
    if((currentstate == "S") and (outcome == 5)): # "Start" state & Dice roll 5 -> "Intermediate"
        return "I"
    elif((currentstate == "I") and (outcome == 5)): # "Intermediate" state & Dice roll 5 -> "Intermediate"
        return "I"
    elif((currentstate == "I") and (outcome == 6)): # "Intermediate" state & Dice roll 6 -> "Finish"
        return "F"
    else:
        return "S"
```

```python
import random
def FSM(update_state): # Passing function as a paramter
    state = "S"
    count = 0 
    while(state != "F"): # Iterate till Final state is reached
        diceroll = random.randint(1,6) 
        state = update_state(state, diceroll) # 
        count = count + 1 # Store each iteration of Dice roll
    return count
```

**Rolling Dice count - 6,6**

```python
FSM(updatestate_6_6)
```

**Rolling Dice count - 5,6**

```python
FSM(updatestate_5_6)
```

**Average of 6,6 in a row for 1000 iterations**

```python
import numpy as np
avg_66 = [FSM(updatestate_6_6) for i in range(1000)] # List that contains each occurences of 6, 6 after 1000 iterations
print(np.mean(avg_66))
```

**Average of 5,6 in a row for 1000 iterations**

```python
import numpy as np
avg_56 = [FSM(updatestate_5_6) for i in range(1000)] # List that contains each occurences of 6, 6 after 1000 iterations
print(np.mean(avg_56))
```
