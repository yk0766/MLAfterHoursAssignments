**1. Find the sum of a list of integers**

```python
lst = [7, 9, 11, 13]
def summ(lst):
    if(len(lst)<1):
        return 0
    else:
        return lst[0] + summ(lst[1:])
print(summ(lst))
```

**2. Find the length of a string**

```python
def ln(s1):
    if(len(s1)==0):
        return 0
    else:
        return 1 + ln(s1[1:])
```

```python
s = "Yash Karande"
print(ln(s))
```

**3. ⁠Find the max integer in a list of integers**

```python
def maxx(lst):
    if(len(lst) < 2):
        return lst[-1]
    elif lst[0] > maxx(lst[1:]):
        max = lst[0]
    else:
        max = maxx(lst[1:])
    return max
        
```

```python
lst1 = [14, 25, 19, 17]
print(maxx(lst1))
```

**4. factorial n** 

```python
def fact(x):
    if(x==1):
        return 1
    else:
        return x * fact(x-1)
```

```python
n = 5
print(fact(n))
```

**5.⁠ Determine if a string is a palindrome**

```python
def palin(p):
    if(len(p)<2):
        return p[-1]
    else:
        c = p[-1] + palin(p[:-1])
        return c
```

```python
z = "civic"
if (z == palin(z)):
    print("Palindrome")
else:
    print("No Palindrome")
```

```python
def palin2(s):
    if(len(s)<=1):
        return True
    else:
        return s[0]== s[-1] and palin2(s[1:-1])
```

```python
z = "noon"
palin2(z)
```

**7th. Fibonanci nth number**

```python
n = 5
def fibo(n):
    if n==0:
        return 0
    elif n==1:
        return 1
    else:
        return fibo(n-1) + fibo(n-2)
```

**8th. Binary Search**

```python
def bs(lst,n):
    i = 0
    j = len(lst)-1
    while(i<=j):
        m = (i+j)//2
        if(n==lst[m]):
            return "Found"
        elif(n>lst[m]):
            i = m+1
        else:
            j = m-1
    return "Not Found"
```

```python
y = [2,3,4,5,6,7,8,9]
x = 7
bs(y,x)
```

```python
def binn(lst,x):
    m = len(lst)//2
    if (len(lst))==0:
        return "Not Found"
    elif(x==lst[m]):
        return "Found"
    elif(x>lst[m]):
        return binn(lst[m+1::],x)
    else:
        return binn(lst[:m],x)
```

```python
lst = [2,3,4,5,6,7,8,9]
x = 1
binn(lst,x)
```

**10th. Count the number of times a character occurs in a string**

```python
def strc(s,c):
    if(len(s)==0):
        return 0
    elif(c==s[0]):
        return 1 + strc(s[1:],c)
    else:
        return strc(s[1:],c)
```

```python
strr = "football"
check = "o"
strc(strr,check)
```
