**2. Write a function that takes a list of integers and calculates the sum of all the odd values. Return the sum and the number of values encountered.** 

```python
def oddsum(numbers):
  summ = 0
  for num in numbers:
    if(num%2!=0):
      summ = summ + num
  return(summ)
```

**3. Determine the max and min of three numbers.**

```python
def maxmin(x,y,z):
    if (x>y):
        max = x
        min = y
    else:
        max = y
        min = x
    if z>x:
        max = z
    elif z<y:
        min = z
    return max, min
```

**6. Calculate the nth Fibonacci number iteratively and recursively.**

```python
def fibo(n):
    fibo = [0,1]
    for i in range(2,n):
      a = fibo[i-2] + fibo[i-1]
      fibo.append(a)
    return fibo[-1]
```

```python
x = int(input("Enter the nth Fibonacci number: "))
fibo(x)
```

**7.Write a function that checks if a given word is a palindrome.**

```python
def palindrome(word):
    i = 0
    j = len(word) - 1
    while(i<j):
        if (word[i]!= word[j]):
            return "Not a Palidnrome word"
        i = i + 1
        j = j - 1
    return "Palidnrome word"
```

```python
s = str(input())
palindrome(s)
```

**10. Check if a Number is an Armstrong Number. An Armstrong number is a number that is the sum of its own digits each raised to the power of the number of digits. For example, 153 = 1^3 + 5^3 + 3^3.**

```python
import math
def arm(x):
    y = len(x)
    summ = 0
    for i in x:
        cub = math.pow(int(i),y)
        summ = summ + cub
    if int(x) == summ:
        return 'Armstrong number'
    else:
        return 'Not an Armstrong number'
```

```python
u = input("Enter a number to check: ")
arm(u)
```

**1. Count Multiples: Given a positive integer n, count how many numbers from 1 up to n are divisible by 3.**

```python
def multi(n): # Define function 'multi' that takes in one paramter 'n'
    c = 0 # Initializing a variable 'c' to store the count of numbers divisible by 3
    for i in range(1,n+1): # Loop through numbers 1 upto 'n' (includes n)
        if(i%3==0): # Check if the number 'i' is divisble by 3
            c = c+1 # If YES then increment the count by 1
    return c # return the final count of numbers divisble by 3
```

```python
multi(10) # calling the function with 10 as the input parameter
```

```python
assert(multi(20)==8)
```

**2. Running Total: You are given a list of numbers representing daily expenses. Starting from zero, add them one by one and report the final total.**



```python
def runtot(lst): # Define function 'runtot' that takes in a list of integers
    total = 0 # Iniialize variable 'total' to store the total
    for i in lst: # Loop that iterates through all the numbers in the list
        total = total + i # Add to previous total
    return total # Return final total 
```

**6. Longest Streak: Given a list of booleans representing whether a task was completed each day, determine the length of the longest consecutive streak of True values.**

```python
def streak(listt):
    c = 0 # Counting the streak of "True"
    maxx = 0 # Storing the longest streak of "True"
    for i in listt:
        if(i == True):
            c = c + 1
            if(c > maxx):
                maxx = c
        else:
            c = 0
    return maxx
    
```

```python
listt = [False, True, True, False, True, True, True, False, True, True, True, True, True, False, True]
streak(listt)
```

**8th. First Repeated Item: Given a list of integers, find the first value that appears more than once when scanning from left to right. If no value repeats, report that fact.**

```python
def repeated(items):
    for i in range(len(items)-1):
        if (items[i] == items[i+1]):
            return items[i]
    return "No repeated items"
```

```python
import random
intgrs = [random.randint(-100, 100) for i in range (20)]
intgrs.sort()
print(intgrs)
repeated(intgrs)
```

**10. Stop on Threshold: You are given a list of numbers and a threshold value. Starting from the beginning of the list, keep adding numbers until the running total becomes strictly greater than the threshold. Report how many numbers were needed.**

```python
def threshold (nums, thresh):
    total = 0
    c = 0
    for i in nums:
        total = total + i
        c = c + 1
        if (total > thresh):
            return c
```

```python
nums = [10, 20, 24, 30, 59]
thresh = 80
threshold(nums, thresh)
```

**11. Prime Counter: Given a positive integer n, count how many prime numbers exist between 2 and n inclusive.**

```python
def prime(n):
    if(n<=1):
        return False
    for i in range(2,n):
        if(n%i==0):
            return False
    return True
```

```python
prime(7)
```

```python
def pc(nums):
    c = 0
    for i in range(2, nums + 1):
        if(prime(i)):
            c = c + 1
    return c
```
