**1. Write a function that takes a list of integers and calculates their sum.**

```python
def summation(lst):
    sum = 0
    for num in lst:
        sum = sum+num
    return sum
```

**2. Write a function that takes a list of integers and calculates the sum of all the odd values. Return the sum and the number of values encountered.** 

```python
def oddsum(numbers):
  summ = 0
  for num in numbers:
    if(num%2!=0):
      summ = summ + num
  return(summ)
```

```python
y = input("Enter the list of integers: " )
numbs = list(map(int,y.split()))
oddsum(numbs)
```

**3. Determine the max and min of three numbers.**

```python
def maxmin(x,y,z):
    if (x>y):
        max = x
        min = y
    else:
        max = y
        min = x
    if z>x:
        max = z
    elif z<y:
        min = z
    return max, min
```

**5. Implement factorial iteratively and recursively.**

```python
def fact(n):
    f = 1
    for i in range(1,n+1):
        f = f*i
    return f
```

```python
x = int(input("Enter the number for factorial: "))
fact(x)
```

**6. Calculate the nth Fibonacci number iteratively and recursively.**

```python
def fibo(n):
    fibo = [0,1]
    for i in range(2,n):
      a = fibo[i-2] + fibo[i-1]
      fibo.append(a)
    return fibo[-1]
```

```python
x = int(input("Enter the nth Fibonacci number: "))
fibo(x)
```

**7.Write a function that checks if a given word is a palindrome.**

```python
def palindrome(word):
    i = 0
    j = len(word) - 1
    while(i<j):
        if (word[i]!= word[j]):
            return "Not a Palidnrome word"
        i = i + 1
        j = j - 1
    return "Palidnrome word"
```

```python
s = str(input())
palindrome(s)
```

**10. Check if a Number is an Armstrong Number. An Armstrong number is a number that is the sum of its own digits each raised to the power of the number of digits. For example, 153 = 1^3 + 5^3 + 3^3.**

```python
import math
def arm(x):
    y = len(x)
    summ = 0
    for i in x:
        cub = math.pow(int(i),y)
        summ = summ + cub
    if int(x) == summ:
        return 'Armstrong number'
    else:
        return 'Not an Armstrong number'
```

```python
u = input("Enter a number to check: ")
arm(u)
```

```python
import math
def reversee(z):
    rev = 0
    while(z>0):
        rem = z%10
        rev = rem*pow(10,lnnum(z)-1) + rev
        z = z//10
    return rev
```

```python
reversee(990)
```

```python
def mag(x):
    if(x>reversee(x)):
        d = x-reversee(x)
        if(lnnum(d)==2):
            d = d*10
    else:
        d = reversee(x)-x
        if(lnnum(d)==2):
            d = d*10
    summ = d + reversee(d)
    return summ
```
