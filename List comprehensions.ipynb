* Create a list containing the squares of odd numbers and half of the even numbers


```python
import random
lst_prnt = [i//2 if(i%2==0) else i*i for i in range(1, 20)]
print(lst_prnt)
```

* Create a list containing 
    * the numbers 0.1, 0.2, 0.3 … upto 1.0
    * The numbers 0.01, 0.02, 0.03 … upto 1.0

```python
lst_od = [i/10 for i in range(1,11)]
print(lst_od)
```

```python
lst_td= [i/100 for i in range(1,11)]
print(lst_td)
```

* Import the math library. You can use `math.sin()` and `math.cos()` to calculate the $\sin$ and $\cos$ of the argument. Create a list of sin and cos of numbers from 0.1 … 1.0 and 0.01 … 1.0
* $\tan$ is easily calculated using $\tan{x} = \sin{x} / \cos{x}$. Create a list of tan() of numbers from 0.1 … 1.0 and 0.01 … 1.0

```python
import math
lst_sin = [math.sin(i) for i in lst_od]
lst_cos = [math.cos(i) for i in lst_od]
lst_tan = [(math.sin(i)) / (math.cos(i)) for i in lst_od]
print("The Sin values: ", lst_sin)
print()
print("The Cos values: ",lst_cos)
print()
print("The Tan value :", lst_tan)
```

```python
import math
lst_sin = [math.sin(i) for i in lst_td]
lst_cos = [math.cos(i) for i in lst_td]
lst_tan = [(math.sin(i)) / (math.cos(i)) for i in lst_td]
print("The Sin values: ", lst_sin)
print()
print("The Cos values: ",lst_cos)
print()
print("The Tan value :", lst_tan)
```
