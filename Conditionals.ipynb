* Check if a number is even or odd

```python
1. ## First Way

import random
def checkeve(n):
    for i in range(n):
        numm = random.randint(1,10000)
        if(numm%2==0):
            print(f" {numm} The number is even")
        else:
            print(f" {numm} The number is odd")
```

```python
x = int(input("Provide the count of numbers"))
checkeve(x)
```

```python
## Second way

def eve1(lst):
    for i in lst:
        if i%2==0:
            print(f" {i} The number is even" )
        else:
            print(f" {i} The number is odd" )
        
```

```python
lst1 = [random.randint(1,10000) for i in range(5)]
eve1(lst1)
```

Triangle Validity and Type: You receive three positive numbers that claim to be the lengths of a triangle’s sides. First decide whether they can form a valid triangle at all. If they can, classify the triangle as equilateral (all sides equal), isosceles (exactly two sides equal), or scalene (all sides different). If they cannot form a triangle, say so.


```python

def tri(x,y,z):
    if((x+y>z) and (y+z>x) and (z+x>y)):
        if(x==y==z):
            return "Equilateral Triangle"
        elif(x!=y!=z):
            return "Scalene Triangle"
        else:
            return "Isosceles Triangle"
    else:
        return "Not a Triangle"
```

```python
tri(-1,0,3)
```

```python
def time(x):
    if((0<=x<=4) or (22<=x<=23)):
        return "night"
    elif(5<=x<=11):
        return "Morning"
    elif(12<=x<=16):
        return "afternoon"
    else:
        return "evening"
```

Interval Relationship: You are given two closed numeric intervals, each described by two endpoints that might be in any order. After interpreting each interval correctly (left endpoint ≤ right endpoint), report whether the first is entirely inside the second, the second is entirely inside the first, the two merely overlap without containment, or they are disjoint.

```python
def inter(x1, x2, y1, y2):
    if(x1<=x2 and y1<=y2):
        if(x1>=y1 and x2<=y2):
            return "First is entirely in second"
        elif(y1>=x1 and y2<=x2):
            return "Second is entirely in First"
        elif((x1<y1) and (x2<y1) or (y1<x1) and (y2<x1)) :
            return "Disjoint"
        else:
            return "Overlap without containment"

    else:
        return "Invalid Input"
        
```

```python
inter(1,3,5,7)
```

```python
import cmath
def quad(a, b, c):
    if(a!=0):
        dis = pow(b,2) - 4*a*c
        if(dis > 0):
            x1 = (- b + cmath.sqrt(dis)) / (2*a)
            x2 = (- b - cmath.sqrt(dis)) / (2*a)
            return x1, x2
        elif(dis == 0):
            return (-b) / (2*a)
        else:
            return "Complex roots"
    else:
        return "Coefficient of a can't be 0"
```
