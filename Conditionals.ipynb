* Check if a number is even or odd

```python
1. ## First Way

import random
def checkeve(n):
    for i in range(n):
        numm = random.randint(1,10000)
        if(numm%2==0):
            print(f" {numm} The number is even")
        else:
            print(f" {numm} The number is odd")
```

```python
x = int(input("Provide the count of numbers"))
checkeve(x)
```

```python
## Second way

def eve1(lst):
    for i in lst:
        if i%2==0:
            print(f" {i} The number is even" )
        else:
            print(f" {i} The number is odd" )
        
```

```python
lst1 = [random.randint(1,10000) for i in range(5)]
eve1(lst1)
```

* Check if a number is positive or negative or zero


```python
def pnz(n):
    if(n==0):
        print(f" {n}, the number is zero")
    elif(n>0):
        print(f" {n}, the number is Positive")
    else:
        print(f" {n}, the number is Negitive")
```

```python
x = int(input("Enter the integer: "))
pnz(x)
```

* Check whether a number is divisible by 3 or 5


```python
def divv1(l):
    if(l%3==0 and l%5==0):
        print(f" {l}, the number is divisible by both 3 or 5")
    elif(l%3==0):
        print(f" {l}, the number is divisible by 3")
    elif(l%5==0):
        print(f" {l}, the number is divisible by 5")
    else:
        print(f" {l}, the number is not divisible by both 3 or 5")
```

```python
divv1(15)
divv2(18)
```

* Print the number of days in a year

```python
def doy(y):
    if ((y%4==0 and y%100!=0) or (y%4==0 and y%100==100 and y%400==0)):
        return 366
    else:
        return 365
```

```python
doy(2020)
```

* Print the number of days in a month

```python
def doyy(y):
    if ((y%4==0 and y%100!=0) or (y%4==0 and y%100==100 and y%400==0)):
        return True
```

```python
def dom(m,y):
    if (m=="Feb"):
        if(doyy(y) == True):
            return 29
        else:
            return 28
    elif(m=="September" or m=="April" or m=="June" or m=="November"):
        return 30
    else:
        return 31
```
